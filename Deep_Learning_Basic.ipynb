### # Linear Regression

```python
## Sinple Linear Regression ##

import numpy as np
import matplotlib.pyplot as plt

x=np.array([2,4,6,8]) # arrary of the hours of study
y=np.array([81,93,91,97]) # array of the test score

# plot the distribution of the data
plt.scatter(x,y)
plt.show()

# initialize the slope and the y-intercept
a=0
b=0

lr=0.03 # learning rate
epochs=2001 # the time of iteration
n=len(x) # length of the data

# gradient decent
for i in range(epochs):
  y_pred=a*x+b
  error=y-y_pred

  a_diff=2/n*sum(-x*error) # partial derivative of MSE with respect to 'a'
  b_diff=2/n*sum(-error) # partial derivative of MSE with respect to 'b'

  # update the value
  a=a-lr*a_diff
  b=b-lr*b_diff

  if i%100==0: # every 100 iteration, the value of a and b will be printed out
    print("epoch=%.f, slope=%.04f, y-intercept=%.04f"%(i,a,b))

y_pred=a*x+b # final prediction

# plot
plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show()
```

```python
## Multiple Linear Regression ##

import numpy as np
import matplotlib.pyplot as plt

x1=np.array([2,4,6,8]) # study hours
x2=np.array([0,4,2,3]) # class hours
y=np.array([81,93,91,97]) # test score

# plot the distribution of the data
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()

# initialize variables
a1=0;a2=0;b=0;
lf=0.01;
epochs=2001;
n=len(x1)

# gradient decent
for i in range(epochs):
  y_pred=a1*x1+a2*x2+b;
  error=y-y_pred;

  #partial derivatives
  a1_diff=2/n*sum(-x1*error)
  a2_diff=2/n*sum(-x2*error)
  b_diff=2/n*sum(-error)

  #update the value
  a1=a1-lf*a1_diff
  a2=a2-lf*a2_diff
  b=b-lf*b_diff

  if i%100==0:
    print("epoch=%.f, a1=%.04f, a2=%.04f, b=%.04f"%(i,a1,a2,b));

y_pred=a1*x1+a2*x2+b;

print("Real score: ",y)
print("Predicted score: ",y_pred)
```

### # Deep Learning (Linear, Logistic)


```python
## Deep learning ##

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#hypothsis H(x)=w*x+b

x=np.array([2,4,6,8]) # study hours
y=np.array([81,93,91,97]) # test score

model=Sequential()

model.add(Dense(1,input_dim=1,activation='linear')); # input_dim: number of independent variable
model.compile(optimizer='sgd',loss='mse'); # sgd: gradient decent
model.fit(x,y,epochs=2000);

#plot
plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()

#test the model
hour=7;
prediction=model.predict(np.array([hour])) # Convert the input to a NumPy array
print("When a student studies %.fhours, that student would be expected to get %.02fscore"%(hour,prediction))
```



### # Multilayer Perceptron



```python
# Solving XOR problem using multilayer perceptron

import numpy as np

# weight and bias
w11=np.array([-2,-2])
w12=np.array([2,2])
w2=np.array([1,1])
b1=3
b2=-1
b3=-1

# perceptron function
def MLP(x,w,b):
  y=np.sum(w*x)+b
  if y<=0: # sigmoid
    return 0
  else:
    return 1

# NAND gate
def NAND(x1,x2):
  return MLP(np.array([x1,x2]),w11,b1)

# OR gate
def OR(x1,x2):
  return MLP(np.array([x1,x2]),w12,b2)

# AND gate
def AND(x1,x2):
  return MLP(np.array([x1,x2]),w2,b3)

# XOR gate
def XOR(x1,x2):
  return AND(NAND(x1,x2),OR(x1,x2))

for x in [(0,0),(1,0),(0,1),(1,1)]:
  y=XOR(x[0],x[1])
  print("input: "+str(x)+ "output: "+str(y))
```

### # K-fold cross validataion

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import pandas as pd

!git clone https://github.com/taehojo/data.github
df=pd.read_csv('./data/sonar3.csv',header=None)

X=df.iloc[:,0:60]
y=df.iloc[:,60]

k=5 # the number of "fold" for the cross validatoin
kfold=KFold(n_splits=k,shuffle=True) # shuffle the data
acc_score=[]

def model_fn():
  model=Sequential()
  model.add(Dense(24,input_dim=60,activation='relu'))
  model.add(Dense(10,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  return model

for train_index, test_index in kfold.split(X):
  X_train, X_test=X.iloc[train_index,:],X.iloc[test_index,:]
  y_train, y_test=y.iloc[train_index],y.iloc[test_index]

  model=model_fn()
  model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
  history=model.fit(X_train,y_train,epochs=200,batch_size=10,verbose=0) # verbose=0; omit presenting the trainning process

  accuracy=model.evaluate(X_test,y_test)[1]
  acc_score.append(accuracy)

# get the average of accuracy
avg_acc_score=sum(acc_score)/k

print('Accuracy:',acc_score)
print('Average of the accuracy:',avg_acc_score)

```

```python
## Improve model by adding validation set

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import pandas as pd

!git clone https://github.com/taehojo/data.git
df=pd.read_csv('./data/wine.csv',header=None)

X=df.iloc[:,0:12]
y=df.iloc[:,12]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True)

model=Sequential()
model.add(Dense(30,input_dim=12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.summary()

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

history=model.fit(X_train,y_train, epochs=50, batch_size=500,validation_split=0.25) # train size x 0.25 = total data x 0.2

score=model.evaluate(X_test,y_test)
print('Test accuracy:',score[1])



```

```python
## Improve model by checking the best performing model of specific epoch

from tensorflow.keras.callbacks import ModelCheckpoint

# Set how the model is saved
modelpath="./data/model/all/{epoch:02d}-{val_accuracy:.4f}.keras"
checkpointer = ModelCheckpoint(filepath=modelpath, verbose=1)

# run the model
history=model.fit(X_train, y_train, epochs=50, batch_size=500, validation_split=0.25,verbose=0,callbacks=[checkpointer])

# print the result
score=model.evaluate(X_test,y_test)
print('Test accuracy:', score[1])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint # Import ModelCheckpoint
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

!git clone https://github.com/taehojo/data.git
df=pd.read_csv('./data/house_train.csv')

df=pd.get_dummies(df)
df=df.fillna(df.mean())
df_corr=df.corr()
df_corr_sort=df_corr.sort_values('SalePrice', ascending=False)
cols_train = ['OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF']

X_train_pre = df[cols_train]
y=df['SalePrice']

X_train, X_test, y_train,y_test = train_test_split(X_train_pre,y, test_size=0.2)

model=Sequential()
model.add(Dense(10,input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1))
model.summary()

# run the model
model.compile(optimizer='adam',loss='mean_squared_error')

early_stopping_callback=EarlyStopping(monitor='val_loss',patience=20)

# name the model
modelpath='./data/model/Ch15-house.keras'

# find the best model and update and share it.

checkpointer=ModelCheckpoint(filepath=modelpath,monitor='val_loss',verbose=0,save_best_only=True)

history=model.fit(X_train,y_train,validation_split=0.25,epochs=2000,batch_size=32,callbacks=[early_stopping_callback, checkpointer])
```

```python
# visualize the outcome prediciton
real_prices=[]
pred_prices=[]
X_num=[]

n_iter=0
Y_prediction = model.predict(X_test).flatten()
for i in range(25):
  real=y_test.iloc[i]
  prediction = Y_prediction[i]
  print("real cost:{:.2f}, predicted cost:{:.2f}".format(real,prediction))
  real_prices.append(real)
  pred_prices.append(prediction)
  n_iter+=1
  X_num.append(n_iter)

plt.plot(X_num,pred_prices,label='predicted price')
plt.plot(X_num,real_prices,label='real price')
plt.legend()
plt.show()
```
